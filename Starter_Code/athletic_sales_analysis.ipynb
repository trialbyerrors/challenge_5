### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
data_2020 = pd.read_csv("Resources/athletic_sales_2020.csv")
data_2021 = pd.read_csv("Resources/athletic_sales_2021.csv")
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
Joined_sails_data_frame=pd.concat([data_2020, data_2021], axis="rows", join="outer")

Joined_sails_data_frame.head(10)
```

```python
# Convert the "invoice_date" to a datetime datatype
Joined_sails_data_frame["dates"] = pd.to_datetime (Joined_sails_data_frame["invoice_date"])
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
total_units_sold = Joined_sails_data_frame.groupby(["region", "state", "city"])["units_sold"].agg(units_sold = "sum") 
# Rename the sum to "Total_Products_Sold".
total_units_sold = total_units_sold.rename(columns={"units_sold  ": "Total_Products_Sold"})
# Show the top 5 results.
total_units_sold.sort_values(by = (["units_sold"]), ascending = False).head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
Total_Products_Sold = pd.pivot_table(Joined_sails_data_frame,
                                     index = ["region", "state", "city"],
                                     values = "units_sold",
                                     aggfunc = "sum")

# Rename the "units_sold" column to "Total_Products_Sold"
Total_Products_Sold = Total_Products_Sold.rename(columns = {"units_sold" : "Total_Products_Sold"})
Total_Products_Sold

# Show the top 5 results.
Total_Products_Sold.sort_values(by = (["Total_Products_Sold"]), ascending = False).head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
Total_Products_Sold = Joined_sails_data_frame.groupby(["region", "state", "city"])["total_sales"].agg(total_sales = "sum")

# Rename the "total_sales" column to "Total Sales"
Total_Products_Sold = Total_Products_Sold.rename(columns = {"total_sales" : "Total Sales"})
Total_Products_Sold
# Show the top 5 results.
Total_Products_Sold.sort_values(by = (["Total Sales"]), ascending = False).head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
Total_Sales = pd.pivot_table(Joined_sails_data_frame,
                             index = ["region", "state", "city"],
                             values = ["total_sales"],
                             aggfunc = "sum")
Total_Sales
# Optional: Rename the "total_sales" column to "Total Sales"
Total_Sales = Total_Sales.rename(columns = {"total_sales" : "Total Sales"})
Total_Sales

# Show the top 5 results.
Total_Products_Sold.sort_values(by = (["Total Sales"]), ascending = False).head()
```

### 4. Determine which Retailer had the Most Sales

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_df  = Joined_sails_data_frame[Joined_sails_data_frame["product"] == "Women's Athletic Footwear"]

womens_df 
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_shoe_sales = womens_df.groupby(["retailer", "region", "state", "city"])["units_sold"].agg(units_sold = "sum")

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_shoe_sales = womens_shoe_sales.rename(columns = {"units_sold" : "Womens_Footwear_Units_Sold"})


# Show the top 5 results.
womens_shoe_sales.sort_values(by = (["Womens_Footwear_Units_Sold"]), ascending = False).head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_shoe_sales = pd.pivot_table(womens_df,
                             index = ["retailer", "region", "state", "city"],
                             values = ["units_sold"],
                             aggfunc = "sum")

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_shoe_sales = womens_shoe_sales.rename(columns = {"units_sold" : "Womens_Footwear_Units_Sold"})


# Show the top 5 results.
womens_shoe_sales.sort_values(by = (["Womens_Footwear_Units_Sold"]), ascending = False).head()
```
